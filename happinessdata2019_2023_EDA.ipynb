```python
#Unique countries considered each year
print("Number of countries in 2019 = ", df_2019['country'].nunique())
print("Number of countries in 2020 = ", df_2020['country'].nunique())
print("Number of countries in 2021 = ", df_2021['country'].nunique())
print("Number of countries in 2022 = ", df_2022['country'].nunique())
print("Number of countries in 2023 = ", df_2023['country'].nunique())
```

```python
#Countries missing each year
countries_2019 = set(df_2019['country'].unique())
countries_2020 = set(df_2020['country'].unique())
countries_2021 = set(df_2021['country'].unique())
countries_2022 = set(df_2022['country'].unique())
countries_2023 = set(df_2023['country'].unique())

missing_countries_2020 = countries_2019 - countries_2020
missing_countries_2021 = countries_2020 - countries_2021
missing_countries_2022 = countries_2021 - countries_2022
missing_countries_2023 = countries_2022 - countries_2023

print("Missing countries in 2020:", missing_countries_2020)
print("Missing countries in 2021:", missing_countries_2021)
print("Missing countries in 2022:", missing_countries_2022)
print("Missing countries in 2023:", missing_countries_2023)
```
